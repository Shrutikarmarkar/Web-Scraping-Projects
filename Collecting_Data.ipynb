### List top 5 characters with maximum number of episodes. Start from the highest.



```python
data = episode_df.loc[:,["name","url","episode"]]
character_dict = {}  # creating a dictionary with keys as character name and value as the number of episodes they appeared in
for i in range (len(data)):
  url = requests.get(data['url'][i])  # episode 1 loop
  df = url.json()
  for j in range (len(df['characters'])):
    characters = df['characters'][j]
    character = requests.get(characters)
    dff = character.json()
    char_name = str(dff['name'])

    if char_name in character_dict:
      character_dict[char_name] +=1
    else:
      character_dict[char_name] = 1

character_dict = sorted(character_dict.items(), key = lambda x: x[1], reverse=True)
print(f'Top 5 characters that appeared in most of the episodes:\n {character_dict[0:5]}')
```

### What is the name of the episode with maximum number of characters?

```python
data = episode_df.loc[:,["name","url","episode"]]
episode_dict = {}  # creating a dictionary with keys as episode name and value as the number of character in it
for i in range (len(data)):
  char_count = 0
  url = requests.get(data['url'][i])  # episode 1 loop
  df = url.json()
  for j in range (len(df['characters'])):
    characters = df['characters'][j]
    character = requests.get(characters)
    dff = character.json()
    char_count +=1
  episode_dict[data['episode'][i]] = char_count

maximum = max(episode_dict,key=episode_dict.get)
print(f"Episode with maximum characters : {maximum}")
```
