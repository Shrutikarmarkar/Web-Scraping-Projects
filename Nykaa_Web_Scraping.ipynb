## Final Code

```python
from numpy import NaN
def get_soup(url):
  try:
    resp = requests.get(url)
  except:
    return None

  if resp.status_code == 200:
    return bs(resp.text)
  else:
    return None

def discount(discounts):
  for dis in discounts:
      if dis.has_attr('class') and 'css-bhhehx' in dis['class']:
          return dis.text

  return 0


def get_details(soup):

  ## Name
  name = soup.find("h1",class_= "css-1gc4x7i").text

  ## rate
  rate = soup.find('div',class_="css-m6n3ou").text

  ## discount
  dis = soup.find_all('span')
  discounts =  discount(dis)

  ## ratings
  ratings = soup.find('div',class_="css-1hvvm95").text.split()[0]

  ## Price
  price = soup.find("span",class_="css-1jczs19").text

  ## reviews
  reviews = soup.find_all('div',class_="css-1hvvm95")
  if len(reviews)==3:
    text = reviews[2].text.split()
    if text[1] == 'reviews':
      reviews = text[0]
  else:
      reviews = NaN

  return name, rate, discounts, ratings, price, reviews



def nykaa(page=3):
  products=[]
  for i in range(1,page+1):
    url = f"https://www.nykaa.com/makeup/face/face-primer/c/233?page_no={i}&sort=popularity&eq=desktop"
    soup = get_soup(url)
    if soup:
      try:
        links = soup.find_all("a",class_="css-qlopj4")
        for link in links:
          primer = "https://www.nykaa.com"+link['href']
          soup = get_soup(primer)

          if soup:
            products.append(get_details(soup))
          else:
            return None

      except:
        print(f"Error reading page {i} . . .")

      time.sleep(1)

  products = pd.DataFrame(products,columns=['Name','Rate','Discounts','Ratings','Price','Reviews'])
  return products
```
